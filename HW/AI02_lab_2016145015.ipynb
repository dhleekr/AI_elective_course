```python
# 2.1
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    
    tmp = x1*w1 + x2*w2 
    
    if tmp <= theta:
        return 0
    else:
        return 1

print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
# 2.4
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    
    tmp = np.sum(w*x) + b

    if tmp <= 0:
        return 0
    else:
        return 1

for xs in [(0,0), (0,1), (1,0), (1,1)]:
    y = NAND(xs[0], xs[1])
    print(str(xs) + ' -> ' + str(y))
```

```python
# 2.5
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    
    tmp = np.sum(w*x) + b

    if tmp <= 0:
        return 0
    else:
        return 1

for xs in [(0,0), (0,1), (1,0), (1,1)]:
    y = OR(xs[0], xs[1])
    print(str(xs) + ' -> ' + str(y))
```
